```python
# Standardization Using Scaler Min = 0, Max = 1

def norm_fun(i):
  x = (i - i.min())/(i.max() - i.min())

  return x
```

```python
data_norm = norm_fun(data)
data = pd.DataFrame(data_norm)
data.describe()
```

```python
# Heat Map

corr_mat = data.corr()
sns.heatmap(data = corr_mat, annot = True)
```

```python
# Spliting the data

X = data.drop(columns = ['Clicked_on_Ad'])
Y = data.Clicked_on_Ad
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0 )
```

```python
# Model Building

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)

# check the accuracy on the training set
model.score(X, Y) # 97% is Preety good Score
```

```python
y_pred = model.predict(x_test)
```

```python
# examine the coefficients
pd.DataFrame(zip(X.columns, np.transpose(model.coef_)))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
## K-fold cross validation
score = cross_val_score(model, X, Y, scoring='accuracy', cv = 10)
print(score)
print(score.mean())
```
