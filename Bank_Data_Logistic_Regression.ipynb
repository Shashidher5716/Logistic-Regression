```python
# Standardization using scaler means max = 1, min = 0

def norm_fun(i):
  x = (i - i.min())/(i.max() - i.min())

  return x
```

```python
data_norm = norm_fun(data)
data = pd.DataFrame(data_norm)
data.describe()
```

```python
# Checking the correaltion 

data.corr()
```

```python
# HeatMap

fig = plt.figure(figsize = (22,10))
sns.heatmap(data.corr(), cmap='viridis', annot = True) 
```

```python
# Spliting the data

X = data.drop(columns = ['y'])
Y = data.y
```

```python
# Using Train Test Split model

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
# Model Building

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs', max_iter=500)
model.fit(x_train, y_train)

# Checking the Accuracy of training set

model.score(X,Y) # 90% is pretty good score
```

```python
y_pred = model.predict(x_test)
```

```python
# examine the coefficients

pd.DataFrame(zip(X.columns, np.transpose(model.coef_)))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
## K-fold cross validation
from sklearn.model_selection import cross_val_score

score = cross_val_score(model, X, Y, scoring='accuracy', cv = 32)
print(score)
print(score.mean())
```
