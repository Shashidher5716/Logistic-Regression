```python
# Dummy Values - Label Encoder

from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()

HD['Sex'] = enc.fit_transform(HD['Sex'])
HD['ChestPainType'] = enc.fit_transform(HD['ChestPainType'])
HD['RestingECG'] = enc.fit_transform(HD['RestingECG'])
HD['ExerciseAngina'] = enc.fit_transform(HD['ExerciseAngina'])
HD['ST_Slope'] = enc.fit_transform(HD['ST_Slope'])
```

```python
# Standardization - Scaling Min = 0, MAx = 1

def norm_fun(i):
    x = (i - i.min())/(i.max() - i.min())
    
    return x
```

```python
data = norm_fun(HD)
HD = pd.DataFrame(data)
HD.describe()
```

```python
# Checking the correlarion

HD.corr()
```

```python
# HeatMap

fig = plt.figure(figsize = (15,10))
sns.heatmap(HD.corr(), cmap = 'viridis', annot = True)
```

```python
# Spliting the data

X = HD.drop('HeartDisease', axis = 1)
Y = HD.HeartDisease
```

```python
# Importing train test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
# Model Building

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)

# Checking the model score

model.score(X, Y)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
# confusion Matrix and accuracy_score

from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_pred, y_test)
print(cm)

accuracy_score(y_pred, y_test)
```
