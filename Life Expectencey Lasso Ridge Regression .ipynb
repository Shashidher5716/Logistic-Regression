```python
# Spliting the data

X = data.drop(labels = ['Country'], axis = 1)
Y = data.Country
```

```python
# Importing the train test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
# Continous to Discreate

X = (X > 0).astype(int)
Y = (Y > 0).astype(int)

x_train = (x_train > 0).astype(int)
y_train = (y_train > 0).astype(int)

x_test = (x_test > 0).astype(int)
y_test = (y_test > 0).astype(int)

data = (data > 0).astype(int)
```

```python
# Model Building

from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)

# Checking the model score for training set

model.score(X, Y)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
# Importing confusion_matrix and Accuracy_score

from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_pred, y_test)
print(cm)

accuracy_score(y_pred, y_test)
```

```python
## K-fold cross validation

from sklearn.model_selection import cross_val_score

score = cross_val_score(model, X, Y, scoring='accuracy', cv = 10)
print(score)
print(score.mean())
```
